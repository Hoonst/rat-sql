```python
from transformers import BertModel, BertTokenizer
```

```python
from transformers import AutoModel, AutoTokenizer
```

```python
bert_name = "bert-large-uncased-whole-word-masking"
electra_name = "google/electra-large-discriminator"

bert_model = AutoModel.from_pretrained(bert_name)
bert_tokenizer = AutoTokenizer.from_pretrained(bert_name)

electra_model = AutoModel.from_pretrained(electra_name)
electra_tokenizer = AutoTokenizer.from_pretrained(electra_name)
```

```python
fake_sentence = "The quick brown fox fake over the lazy dog"


fake_tokens = electra_tokenizer.tokenize(fake_sentence)
fake_inputs = electra_tokenizer.encode(fake_sentence, return_tensors="pt")
discriminator_outputs = electra_model(fake_inputs)
```

```python
fake_sentence = "The quick brown fox fake over the lazy dog"


fake_tokens = bert_tokenizer.tokenize(fake_sentence)
fake_inputs = bert_tokenizer.encode(fake_sentence, return_tensors="pt")
bert_outputs = bert_model(fake_inputs)
```
